## Barplot

**Dataset**

price: price in dollars ($326-$18,823)

carat: weight (0.2-5.01)

cut: quality (Fair, Good, Very Good, Premium, Ideal)

color: color (from J (worst) to D (best))

clarity: cleanliness, clarity (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x: length in mm (0-10.74)

y: width in mm (0-58.9)

z: depth in mm (0-31.8)

depth: percentage of total depth = z / mean(x, y) = 2*z / (x + y) (43-79)

table: width relative to the widest point of the diamond (43-95)
